```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../dataset/final_life_expectancy.csv')
```

```python
df
```

### Splitting the dataset

```python
y = df['Life expectancy']
X = df.drop('Life expectancy', axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lr_classifier = LinearRegression()
lr_classifier.fit(X_train, y_train)
lr_predictions = lr_classifier.predict(X_test)
```

```python
print("Mean Absolute Error:",mean_absolute_error(lr_predictions, y_test))
print("Mean Squared Error:", mean_squared_error(lr_predictions, y_test))
```

```python
plt.scatter(lr_predictions, y_test)
m, b = np.polyfit(lr_predictions, y_test, 1)
plt.plot(lr_predictions, m*lr_predictions + b)
```

## Ridge Regression

```python
from sklearn.linear_model import Ridge
```

```python
rd_classifier = Ridge()
rd_classifier.fit(X_train, y_train)
rd_predictions = rd_classifier.predict(X_test)
```

```python
print("Mean Absolute Error:",mean_absolute_error(rd_predictions, y_test))
print("Mean Squared Error:", mean_squared_error(rd_predictions, y_test))
```

```python
plt.scatter(rd_predictions, y_test)
m, b = np.polyfit(rd_predictions, y_test, 1)
plt.plot(rd_predictions, m*rd_predictions + b)
```

## Bayesian Linear Regression

```python
from sklearn.linear_model import BayesianRidge
```

```python
blr_classifier = BayesianRidge()
blr_classifier.fit(X_train, y_train)
blr_predictions = blr_classifier.predict(X_test)
```

```python
print("Mean Absolute Error:",mean_absolute_error(blr_predictions, y_test))
print("Mean Squared Error:", mean_squared_error(blr_predictions, y_test))
```

## SVM

```python
from sklearn import svm
```

```python
svm_classifier = svm.SVR()
svm_classifier.fit(X_train, y_train)
svm_predictions = svm_classifier.predict(X_test)
```

```python
print("Mean Absolute Error:",mean_absolute_error(svm_predictions, y_test))
print("Mean Squared Error:", mean_squared_error(svm_predictions, y_test))
```

```python
plt.scatter(svm_predictions, y_test)
m, b = np.polyfit(svm_predictions, y_test, 1)
plt.plot(svm_predictions, m*svm_predictions + b)
```

### Hence the best model for this dataset is Linear Regression with Mean Absolute Error: 2.565
